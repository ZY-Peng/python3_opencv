```python
# 2D 卷积 
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/home/pzy/python/images/29.jpg')
```

```python
kernel=np.ones((5,5),np.float32)/50
dst=cv2.filter2D(img,-1,kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#图像模糊（图像平滑） 
  1# 平均 cv2.blur()
blur=cv2.blur(img,(25,25))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
  2#高斯模糊 
blur=cv2.GaussianBlur(img,(15,15),0)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('GaussianBlur')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#中值模糊 
blur=cv2.medianBlur(img,15) #卷积核的大小也应该是一个奇数。 
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('medianBlur')
plt.xticks([]),plt.yticks([])
plt.show()
print(blur.shape)
```

```python
# 双边滤波 
blur=cv2.bilateralFilter(img,9,105,105)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]),plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('bilateralFilter')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#自定义模糊 
def custom(image):
    kernel=np.ones([5,5],np.float32)/25
    dst=cv2.filter2D(image,-1,kernel=kernel)
    cv2.imshow('custom',dst)
custom(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
