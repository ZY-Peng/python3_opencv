```python
import cv2
import numpy as np
img1=cv2.imread('/home/pzy/python/images/qipu02.jpg')
img2=cv2.imread('/home/pzy/python/images/qipu.jpg')
print(img1.shape)
print(img2.shape)
```

```python
from matplotlib import pyplot as plt
img2=cv2.resize(img2,(270,300),interpolation=cv2.INTER_LINEAR)
print(img2.shape)
dst=cv2.addWeighted(img1,0.9,img2,0.1,0)
font=cv2.FONT_HERSHEY_SIMPLEX
#各参数依次是：图片，添加的文字，左上角坐标，字体，字体大小，颜色，字体粗细
cv2.putText(dst,'opencv',(40,200),font,2,(0,0,0),2)
plt.subplot(121),plt.imshow(dst,'gray'),plt.title('mixed')

plt.show()
```

```python
bitwiseand=cv2.bitwise_and(img1,img2)

bitwiseor=cv2.bitwise_or(img1,img2)

bitwisexor=cv2.bitwise_xor(img1,img2)

bitwisenot=cv2.bitwise_not(img1,img2)

plt.subplot(221),plt.imshow(bitwiseand,'gray'),plt.title('and')
plt.subplot(222),plt.imshow(bitwiseor,'gray'),plt.title('or')
plt.subplot(223),plt.imshow(bitwisexor,'gray'),plt.title('xor')
plt.subplot(224),plt.imshow(bitwisenot,'gray'),plt.title('not')

plt.show()
```
